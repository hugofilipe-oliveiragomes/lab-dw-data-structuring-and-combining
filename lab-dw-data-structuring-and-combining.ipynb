In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
url4 = ("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv")
df4 = pd.read_csv(url4)
```

```python
df123.reset_index(drop=True, inplace=True)
df4.reset_index(drop=True, inplace=True)
```

```python
df1234 = pd.concat([df123, df4], axis=0)
display(df1234)
```

```python
df1234['customer_lifetime_value'] = df1234['customer_lifetime_value'].str.replace('%', '', regex=True)
display(df1234)
```

```python
df1234['customer_lifetime_value'] = df1234['customer_lifetime_value'].str.replace('%', '', regex=True)
df1234['customer_lifetime_value'] = pd.to_numeric(df1234['customer_lifetime_value'], errors='coerce')
```

```python
df1234['customer_lifetime_value'] = df1234['customer_lifetime_value'].fillna(0).astype(int)
```

```python
df1234.dropna(subset=['customer_lifetime_value'], inplace=True)
df1234['customer_lifetime_value'] = df1234['customer_lifetime_value'].astype(int)
display (df1234)
```

```python
df1234['gender'] = df1234['gender'].astype(str).str.lower()

gender_map = {
    'f': 'Female',
    'm': 'Male',
    'female': 'Female',
    'male': 'Male',
    'femal': 'Female',
    'nan': 'Unknown'  # Explicitly handle 'nan' as a string
}

df1234['gender_mapped'] = df1234['gender'].map(gender_map)

print(df1234[['gender', 'gender_mapped']].head())
```
